# Time series forecasting

Packages

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

Import data from excel file

```python
data = pd.read_excel("caw1.xlsx")
```

```python
Year = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012]
```

# Forecasting rape crime

```python
rape_case = data.iloc[35:36,2:].values
rape_case = rape_case.flatten()
rape_case = rape_case.reshape(12, 1)
```

```python
temp1 = rape_case.flatten()
```

```python
plt.plot(Year, rape_case)
plt.xlabel("Year")
plt.ylabel("No of Rape cases")
plt.title("Number of Rape cases all over India")
plt.show()
```

```python
model1 = ARIMA(rape_case, order=(1, 1, 1))
model_fit = model1.fit()

forecast1 = model_fit.predict(13, 17)
forecast1 = forecast1.round(0)
```

```python
temp = 2013
for i in range(0,5):
    print("Predicted cases for ",temp,"is ",forecast1[i])
    temp = temp + 1
```

```python
pred_year = [2013,2014,2015,2016,2017]
final_year = []
final_pre1 = []
for i in range(0,len(Year)):
    final_year.append(Year[i])
for i in range(0, len(pred_year)):
    final_year.append(pred_year[i])
    
    
for i in range(0,len(temp1)):
    final_pre1.append(temp1[i])
for i in range(0, len(forecast1)):
    final_pre1.append(forecast1[i])
```

```python
plt.plot(final_year, final_pre1)
plt.xlabel("Year")
plt.ylabel("No of Rape cases(Predicted)")
plt.title("Number of Rape cases all over India(Predicted)")
plt.show()
```
